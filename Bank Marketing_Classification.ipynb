# Replacing strings with numbers

```python
#Function for dictonary to store coding info about varaibles
inf0={}
def code_map(le,name):
    global info
    job_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    info[name]=job_name_mapping
    print(name,' :varaiable successfully coded')       
```

```python
data.head(1)
```

```python
#import the necessary module
from sklearn import preprocessing
# create the Labelencoder object
le = preprocessing.LabelEncoder()
#convert the categorical columns into numeric
data['job']=le.fit_transform(data['job'])
code_map(le,'job')

data['marital']=le.fit_transform(data['marital'])
code_map(le,'marital')

data['education']=le.fit_transform(data['education'])
code_map(le,'education')

data['default']=le.fit_transform(data['default'])
code_map(le,'default')

data['housing']=le.fit_transform(data['housing'])
code_map(le,'housing')

data['loan']=le.fit_transform(data['loan'])
code_map(le,'loan')

data['poutcome']=le.fit_transform(data['poutcome'])
code_map(le,'poutcome')

data.head()
```

```python
info
```

```python
#Converting month name into month number
from time import strptime
data['month']=list(map(lambda x:strptime(x,'%b').tm_mon,data['month']))
data.head()
```

# Splitting the dataset into train and test sets

```python
X = data.iloc[:,0:-1]
y = data.iloc[:,-1]
```

# Train Dataset

```python

from sklearn.model_selection import train_test_split
X_test,X_train,y_test,y_train = train_test_split(X, y, test_size =0.3 , random_state =100 )
```

# **************** Model Building *********************

# Binary Logisitic Classification Algorithm

```python
#Fitting a logistic regression model to the train set
from sklearn.linear_model import LogisticRegression
LG=LogisticRegression()
LG.fit(X_train,y_train)
```

```python
y_pred =LG.predict(X_test)
```

```python
#Making predictions and evaluating performance
#Accuracy Score
from sklearn.metrics import accuracy_score
acc_score=accuracy_score(y_test,y_pred)
print('accuracy_score is ',acc_score)
```

```python
#Confusion matrix
from sklearn import metrics
cnf_mat=metrics.confusion_matrix(y_test,y_pred)
print(cnf_mat)
```

```python
#Testing by sending parameters into model
ip=[30,10,1,0,0,1787,0,0,19,10,79,1,-1,0,3]
LG.predict([ip])
```

```python
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier(n_estimators=200, random_state=1)
RFC.fit(X_train, y_train)
y_pred = RFC.predict(X_test)
print('Accuracy',metrics.accuracy_score(y_test,y_pred))
```

From above models Random Forest and Gradient Boost Models gives better accuracy . In these two Random Forest gives more better accuracy
